```python
# 라이브러리 불러오기
import urllib.request 
import json 
import pandas as pd 
```

```python
# 인증키와 주소
key = '7650532331'
start = 1
end = 52
use_dt = 2022
result = []

for nums in range(1, 53):
    url = f'http://data.ex.co.kr/openapi/specialAnal/intercityLeadTime?key=7650532331&type=json&iYear=2021&numOfRows=1000&pageNo={nums}'

    response = urllib.request.urlopen(url) 
    json_str = response.read().decode('utf-8')
    json_object = json.loads(json_str)

    result += json_object['intercityLeadTimeLists']

data = pd.json_normalize(result)
```

```python
drop_cols = ['iYear', 'iStUnitCode', 'iEdUnitCode', 'message', 'code']
data.drop(drop_cols, axis=1, inplace=True)
```

```python
data.columns = ['시간', '연도', '연휴이름', '연휴이름코드',
       '연휴기간', '연휴기간코드', '출발지역', '출발지역코드',
       '도착지역', '도착지역코드', '방향', '방향코드', '소요시간']
```

```python
data['소요시간'] = data['소요시간'].astype(float)
data['소요시간'] = round(data['소요시간'], 3)
```
