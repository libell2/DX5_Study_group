# 코로나 이전(2017~2019) 데이터

- 데이터 불러오기 및 전처리

```python
# 불러오기
data = pd.read_csv('thanksgiving.csv')

# 확인
data.head()
```

```python
# 열 제거
drop_cols = ['출발지역코드', '도착지역코드', '연휴기간코드', '연휴이름코드', '방향코드', 'Unnamed: 0']
data.drop(drop_cols, axis=1, inplace=True)
data.head()
```

```python
# 추석 데이터만 보기
data = data.loc[data['연휴이름'] == '추석']
data['연휴이름'].value_counts()
```

```python
# 소요시간 실수화
data['소요시간'] = data['소요시간'].astype(float)

data.info()
```

```python
data_bk = data.copy()
```

```python
# 코로나 이전 데이터만 활용
data = data.loc[data['연도'].isin([2017, 2018, 2019])]
data.head()
```

```python
data.shape
```

```python
# 열 제거
drop_cols = ['연도', '연휴이름', '출발지역', '도착지역', '방향']
data.drop(drop_cols, axis=1, inplace=True)
data.head()
```

```python
# 결측치 제거
data.dropna(axis=0, inplace=True)
```

```python
# 불러오기
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import *
```

```python
# 데이터 나누기
target = '소요시간'
x = data.drop(target, axis=1)
y = data[target]
```

```python
# 가변수화
dumm_cols = ['시간', '연휴기간']
x = pd.get_dummies(x, columns=dumm_cols, drop_first=False)
x
```

```python
# 데이터 나누기
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# 교차검증
model = LinearRegression()
cv_score = cross_val_score(model, x_train, y_train, cv=10)

print('평균:', cv_score.mean())
```

```python
model = DecisionTreeRegressor(max_depth=4)
cv_score = cross_val_score(model, x_train, y_train, cv=10)

print('평균:', cv_score.mean())
```

```python
model = KNeighborsRegressor(n_neighbors=44)
cv_score = cross_val_score(model, x_train, y_train, cv=10)

print('평균:', cv_score.mean())
```

```python
# 평가
model = LinearRegression()

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
